```
from newspaper import Article
import nltk
from gtts import gTTS
import os
```

```
import newspaper
newspaper.languages()
```

```
#article = Article('https://www.bbc.com/news/world-us-canada-56204449')
url = 'http://fox13now.com/2013/12/30/new-year-new-laws-obamacare-pot-guns-and-drones/'
article = Article(url)
article.download()
article.parse()
print(article.text)
nltk.download('punkt')
```

```
article.nlp()
mytext = article.text
#print(mytext)
language = 'en'
myobj = gTTS(text=mytext, lang=language, slow=False)
myobj.save("read_articlea1.mp3")
#os.system("start read_articlea1.mp3")
```
